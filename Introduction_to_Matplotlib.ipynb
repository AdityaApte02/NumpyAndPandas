# Plotting figures using Numpy Arrays

## We are going to plot the following
1. Line Plot
2. Scatter Plot
3. Bar Plot
4. Histogram
5. Subplot


```python
import numpy as np
```

```python
x=np.linspace(0,20,100)
```

```python
x
```

```python
fig,ax=plt.subplots()
ax.plot(x, x ** 3, color='black', label='x**3');
ax.plot(x, x**1.5, color='cyan',label='x**1.5');
ax.set_xlabel('x')
ax.set_ylabel('x raised to a power')
ax.legend();
```

```python
fig,ax=plt.subplots()
ax.scatter(x, np.exp(x), color='m');
```

```python
# Another scatter plot
fig,ax=plt.subplots()
ax.scatter(x,np.log2(x));
```

```python
# Bar Plot
fig, ax=plt.subplots()
butter_prices={'Almond Butter':10, 'Peanut Butter':8, 'Cashew Butter':12}
ax.bar(butter_prices.keys(),butter_prices.values());
ax.set(xlabel='Butter Types', ylabel='Price in dollars per kg', title='Butter Info')
```

```python
# Another scatter plot example
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
cmap='viridis')
plt.colorbar(); # show color scale
```

```python
# Horizontal Bar Graph
fig, ax=plt.subplots()
ax.barh(list(butter_prices.keys()), list(butter_prices.values()));
ax.set(ylabel='Butter Types', xlabel='Price in dollars per kg', title='Butter Info');
```

```python
# Make data for Histogram and plot it
np.random.seed(23)
x=np.random.randn(100)
fig,ax=plt.subplots()
ax.hist(x);
```

```python
# Subplot Option 1

fig, ax=plt.subplots(nrows=2, ncols=2, figsize=(10,10))

ax[0][0].plot(x,x/2);
ax[0][1].scatter(np.random.rand(10), np.random.rand(10));
ax[1][0].bar(butter_prices.keys(), butter_prices.values());
ax[1][1].hist(x);
```

```python
# Subplot Option 2
fig, ((ax1,ax2),(ax3, ax4))=plt.subplots(nrows=2,ncols=2, figsize=(10,10))

ax1.plot(x,x/2);
ax2.scatter(np.random.rand(10), np.random.rand(10));
ax3.bar(butter_prices.keys(), butter_prices.values());
ax4.hist(x);
```

```python
# Heart Disease DataFrame
heart_disease=pd.read_csv('./heart-disease.csv')
heart_disease[:10]
```

```python
heart_disease.plot.hist(figsize=(10,20), subplots=True);
```

```python
over_50=heart_disease[heart_disease['age'] > 50]
```

```python
over_50
```

```python
len(over_50)
```

```python
# Scatter plot using the pyplot method
over_50.plot(kind='scatter', x='age', y='chol', c='target', figsize=(10,6));
```

```python
# Using the OO method mixed with the pyplot method
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);
```

```python
# Scatter plot using OO method from scratch

fig, ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50['age'], y=over_50['chol'],c=over_50['target'], cmap='winter');


ax.set(xlabel='Age', ylabel='Cholestrol', title='Age and Cholestrol Plot');
ax.legend(*scatter.legend_elements(), title="Legend");
```

```python
# Subplot of age,cholestrol and thalach
plt.style.use('seaborn-white')
fig, (ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(12,12), sharex=True)
scatter=ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'], cmap='autumn');
ax0.set(title='Heart Disease and Cholestrol', ylabel='Cholestrol Level');
ax0.legend(*scatter.legend_elements(), title='Legend');
ax0.axhline(y=over_50['chol'].mean(), linestyle='dotted', color='green');


scatter1=ax1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'], cmap='summer')
ax1.set(title='Heart Disease and Thalach', xlabel='Age', ylabel='Thalach')
ax1.legend(*scatter1.legend_elements(), title='Legend');
ax1.axhline(y=over_50['thalach'].mean(), linestyle='--', color='purple');

fig.suptitle('Heart Disease Analysis', fontsize=20, fontweight='bold');
```
